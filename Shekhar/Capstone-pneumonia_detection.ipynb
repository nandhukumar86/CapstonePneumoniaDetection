```python
def disp_xray(patient_id):
    train_example1 =  pydicom.dcmread(train_files_path+'/'+train_file_names[patient_id])
    print(train_file_names[patient_id])
    plt.imshow(train_example1.pixel_array,cmap='gray')
    if(train_labels_df['Target'][patient_id] == 1):
        plt.gca().add_patch(Rectangle((train_labels_df['x'][patient_id],train_labels_df['x'][patient_id]),train_labels_df['width'][patient_id],train_labels_df['height'][patient_id],linewidth=2,edgecolor='r',facecolor='none'))       
```
